```python
import pandas as pd
import numpy as np
import datetime
from sklearn import preprocessing
import random
```

```python
#Open the initial file and drop unwanted features
df =pd.read_csv("./traffic_accidents.csv",low_memory=False)
df=df.drop(columns=['OBJECTID_1','INCIDENT_ID','OFFENSE_ID','OFFENSE_CODE','OFFENSE_CODE_EXTENSION','TOP_TRAFFIC_ACCIDENT_OFFENSE','LAST_OCCURRENCE_DATE','REPORTED_DATE','INCIDENT_ADDRESS','GEO_X','GEO_Y','DISTRICT_ID','PRECINCT_ID','NEIGHBORHOOD_ID','HARMFUL_EVENT_SEQ_1', 'HARMFUL_EVENT_SEQ_2', 'HARMFUL_EVENT_SEQ_3','TU2_VEHICLE_TYPE', 'TU2_TRAVEL_DIRECTION', 'TU2_VEHICLE_MOVEMENT','TU2_DRIVER_ACTION', 'TU2_DRIVER_HUMANCONTRIBFACTOR','TU2_PEDESTRIAN_ACTION', 'SERIOUSLY_INJURED', 'FATALITIES','FATALITY_MODE_1', 'FATALITY_MODE_2', 'SERIOUSLY_INJURED_MODE_1','SERIOUSLY_INJURED_MODE_2','TU1_VEHICLE_MOVEMENT' ,'TU1_DRIVER_ACTION','TU1_PEDESTRIAN_ACTION'])
df=df.dropna(how='any')
df.to_csv("Processed.csv",index=False)
```

```python
#From the features, get day, month, hour and create its own column
df=pd.read_csv("./Processed.csv")
data=df['FIRST_OCCURRENCE_DATE']
weekday=[]
hour=[]
month=[]
day=[]
year=[]

for item in data.iteritems():
    items= item[1].split(" ")
    hour.append(items[1].split(":")[0])
    
    date = datetime.datetime.strptime(items[0],"%Y-%m-%d")
    weekday.append(date.weekday())
    month.append(date.month)
    day.append(date.day)
    year.append(date.year)
    
df['Weekday']=weekday
df['hour']=hour
df['month']=month
df['day']=day
df['year']=year
df=df.drop(columns=['FIRST_OCCURRENCE_DATE'])
```

```python
# process the features to have numerical values
le=preprocessing.LabelEncoder()

index= ['ROAD_LOCATION','ROAD_DESCRIPTION', 'ROAD_CONTOUR', 'ROAD_CONDITION', 'LIGHT_CONDITION','TU1_VEHICLE_TYPE', 'TU1_TRAVEL_DIRECTION','TU1_DRIVER_HUMANCONTRIBFACTOR']

for i in index:
    df[i]=np.where(df[i]=="  ",None,df[i])
    df[i]=np.where(df[i]=="UNDER INVESTIGATION",None,df[i])
df=df.dropna(how='any')
    
for i in index:
    prop=df[i]
    le.fit(prop)
    print(le.classes_)
    a=[i for i in range(len(le.classes_))]
    prop=le.transform(prop)
#     print(a)
#     print(le.inverse_transform(a))
    df[i]=prop
    
df['ISACCIDENT']=[1 for x in range(len(df))]
df.to_csv("Processed.csv",index=False)
```
